```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set_theme()
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('EDA-on-IMDB-Movies-Dataset/2.Cleaning/cleaned.csv')
```

```python
df.head()
```

```python
df.drop(['Unnamed: 0','index'], axis=1, inplace=True)
```

```python
df.head()
```

# Top rated movies on IMDB

```python
# I set as top movies on imdb the ones with score equal or more than 8.8
df_top_rated = df[df['IMDB_Rating'] >= 8.8].sort_values(by='IMDB_Rating', ascending=False)
```

```python
df_top_rated
```

```python
# Plotting with Seaborn
plt.figure(figsize=(10, 5))
ax =sns.barplot(
        data=df_top_rated,
        x="IMDB_Rating",
        y="Movie_Title",
        hue="Movie_Title",          # Assign `x` variable to `hue`
        palette="Blues_d",   # Sequential color palette
        dodge=False,         # Prevent separation of bars
        legend=False         # Remove legend since `hue` is the same as `x`
    )
for i in ax.containers:
    ax.bar_label(i,)

# Customizing the plot
plt.title("Top IMDb Ratings of Movies", fontsize=16)
plt.xlabel("IMDb Rating", fontsize=10)
plt.ylabel("Movie Title", fontsize=10)
plt.xticks(rotation=90)  # Rotate movie names for better readability
plt.tight_layout()

# Show the plot
plt.show()
```

# MetaScore of Top IMdb movies

What is the difference between Metascore and IMDb rating?
- Metacritic and Rotten Tomatoes are review aggregate sites that collect critic reviews, while IMDB is a movie information site that has a popularly-cited audience rating.

```python
# Plotting with Seaborn
plt.figure(figsize=(10, 5))
ax =sns.barplot(
        data=df_top_rated,
        x="Meta_score",
        y="Movie_Title",
        hue="Movie_Title",          # Assign `x` variable to `hue`
        palette="Purples_d",   # Sequential color palette
        dodge=False,         # Prevent separation of bars
        legend=False         # Remove legend since `hue` is the same as `x`
    )
for i in ax.containers:
    ax.bar_label(i,)

# Customizing the plot
plt.title("Top Metascore Movies", fontsize=16)
plt.xlabel("Metascore", fontsize=10)
plt.ylabel("Movie Title", fontsize=10)
plt.xticks(rotation=90)  # Rotate movie names for better readability
plt.tight_layout()

# Show the plot
plt.show()
```
